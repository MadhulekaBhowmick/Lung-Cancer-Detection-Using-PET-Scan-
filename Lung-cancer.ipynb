```python
import os
import time
import numpy as np
import torch as T
import torch.nn as nn
import torchvision
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from torchvision import models, transforms
from torch.utils.data import DataLoader, random_split
from torch.optim import Adam
from sklearn.metrics import confusion_matrix, classification_report

device = T.device("cuda" if T.cuda.is_available() else "cpu")

params = {
    'batch_size': 32,
    'epochs': 30,
    'lr': 0.0003,
    'image_size': 224,  
    'mean': [0.485, 0.456, 0.406],
    'std': [0.229, 0.224, 0.225],
    'device': device,
    'checkpoint_dir': "/kaggle/working",
    'data_dir': "/kaggle/input/pet-scans-of-lungs-98-accuracy/gdrive/MyDrive/Projects/Lung-Cancer/Data",
    'dataset_name': "Lung-PET-CT-Dx",
    'disease_names': {
        'A': 'Adenocarcinoma',
        'B': 'Small Cell Carcinoma',
        'E': 'Large Cell Carcinoma',
        'G': 'Squamous Cell Carcinoma',
    },
    'patience': 5
}

os.makedirs(params['checkpoint_dir'], exist_ok=True)

data_transforms = transforms.Compose([
    transforms.Resize(params['image_size']),
    transforms.CenterCrop(params['image_size']),  
    transforms.ToTensor(),
    transforms.Normalize(params['mean'], params['std']),
])

dataset = torchvision.datasets.ImageFolder(params['data_dir'], transform=data_transforms)

# Splitting dataset
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

train, val, test = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train, batch_size=params['batch_size'], shuffle=True, num_workers=2)
val_loader = DataLoader(val, batch_size=params['batch_size'], shuffle=False, num_workers=2)
test_loader = DataLoader(test, batch_size=params['batch_size'], shuffle=False, num_workers=2)

class SwinCancerDetector(nn.Module):
    def __init__(self, base_model, n_classes, lr, name):
        super(SwinCancerDetector, self).__init__()
        
        self.path = os.path.join(params['checkpoint_dir'], name + '.pth')
        self.base_model = base_model

        # Replace Swin's original head with custom classifier
        self.in_features = self.base_model.head.in_features
        self.base_model.head = nn.Identity()  # Remove classification head

        self.classifier = nn.Sequential(
            nn.Linear(self.in_features, 512),
            nn.GELU(),  # Changed from ReLU for better transformer compatibility
            nn.Dropout(0.3),
            nn.LayerNorm(512),  # Added normalization
            nn.Linear(512, 256),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.LayerNorm(256),
            nn.Linear(256, n_classes),
        )

        self.criterion = nn.CrossEntropyLoss()
        self.optimizer = Adam(self.parameters(), lr)

    def forward(self, inputs):
        features = self.base_model(inputs)
        logits = self.classifier(features)
        return logits

base = models.swin_t(weights=models.Swin_T_Weights.DEFAULT)

for param in base.parameters():
    param.requires_grad = False

for block in base.features[2:]:
    for param in block.parameters():
        param.requires_grad = True

CancerDetector = SwinCancerDetector(base, len(params['disease_names']), params['lr'], name='kaggle-swin-t')
Model = CancerDetector.to(params['device'])

# ------------------------ Training Function ------------------------
def fit(Model, loader, train=True):
    total_loss = 0
    total_correct = 0
    all_preds, all_labels = [], []

    Model.train() if train else Model.eval()

    scaler = T.amp.GradScaler("cuda")

    for images, labels in tqdm(loader):
        images, labels = images.to(device), labels.to(device)

        with T.set_grad_enabled(train):
            with T.amp.autocast("cuda"):
                outputs = Model(images)
                loss = Model.criterion(outputs, labels)

            if train:
                scaler.scale(loss).backward()
                scaler.step(Model.optimizer)
                scaler.update()
                Model.optimizer.zero_grad()

        total_loss += loss.item() * images.size(0)
        _, preds = T.max(outputs, 1)
        total_correct += T.sum(preds == labels.data)
        
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    epoch_loss = total_loss / len(loader.dataset)
    epoch_acc = total_correct.double() / len(loader.dataset)
    
    return epoch_loss, epoch_acc.item(), all_preds, all_labels

# ------------------------ Training Loop ------------------------
history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
best_val_loss = float('inf')
early_stop_counter = 0
start_time = time.time()

for epoch in range(params['epochs']):
    print(f'\nEpoch {epoch+1}/{params["epochs"]}')
    
    # Training phase
    train_loss, train_acc, _, _ = fit(Model, train_loader, train=True)
    # Validation phase
    val_loss, val_acc, val_preds, val_labels = fit(Model, val_loader, train=False)
    
    # Update history
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    
    print(f'Train Loss: {train_loss:.4f} | Acc: {train_acc:.4f}')
    print(f'Val Loss: {val_loss:.4f} | Acc: {val_acc:.4f}')
    
    # Early stopping check
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        T.save(Model.state_dict(), Model.path)
        early_stop_counter = 0
        print(' Checkpoint saved')
    else:
        early_stop_counter += 1
        print(f' Early stopping counter: {early_stop_counter}/{params["patience"]}')
        
    if early_stop_counter >= params['patience']:
        print(' Early stopping triggered!')
        break

print(f'Total training time: {(time.time()-start_time)/60:.2f} minutes')

# ------------------------ Evaluation ------------------------
# Load best model
Model.load_state_dict(T.load(Model.path))

# Test evaluation
test_loss, test_acc, test_preds, test_labels = fit(Model, test_loader, train=False)
print(f'\nTest Loss: {test_loss:.4f} | Test Acc: {test_acc:.4f}')

# Confusion matrix
cm = confusion_matrix(test_labels, test_preds)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=params['disease_names'].values(),
            yticklabels=params['disease_names'].values())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# Classification report
print(classification_report(test_labels, test_preds,
                            target_names=params['disease_names'].values()))

# Plot training history
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Val Loss')
plt.legend()
plt.title('Loss Curves')

plt.subplot(1,2,2)
plt.plot(history['train_acc'], label='Train Acc')
plt.plot(history['val_acc'], label='Val Acc')
plt.legend()
plt.title('Accuracy Curves')
plt.show()
```

```python
#DS latest
import os
import time
import torch as T
import torch.nn as nn
import torchvision
from torchvision import models, transforms
from torch.utils.data import DataLoader, random_split
from tqdm import tqdm

# ------------------------ Kaggle-Specific Setup ------------------------
#  Verify GPU availability
print(f"PyTorch version: {T.__version__}")
print(f"CUDA available: {T.cuda.is_available()}")
print(f"CUDA device count: {T.cuda.device_count()}")
print(f"Current device: {T.cuda.current_device()}")
print(f"Device name: {T.cuda.get_device_name(0)}")

# ------------------------ Configuration ------------------------
class Config:
    # Hardware
    device = T.device("cuda" if T.cuda.is_available() else "cpu")
    num_workers = 2  #  Optimal for Kaggle
    
    # Model
    model_name = "swin_tiny"
    pretrained = True
    num_classes = 4
    
    # Training
    image_size = 224
    batch_size = 32  #  P100 can handle 16 with Swin-T
    epochs = 30
    lr = 3e-4
    patience = 5
    
    # Paths
    data_dir = "/kaggle/input/pet-scans-of-lungs-98-accuracy/gdrive/MyDrive/Projects/Lung-Cancer/Data"
    save_dir = "/kaggle/working/checkpoints"
    save_name = f"{model_name}_lung_cancer.pth"
    
    # Augmentation
    mean = [0.485, 0.456, 0.406]
    std = [0.229, 0.224, 0.225]
    
config = Config()

# ------------------------ Data Preparation ------------------------
#  Kaggle-optimized transforms
train_transforms = transforms.Compose([
    transforms.Resize((config.image_size, config.image_size)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    transforms.Normalize(config.mean, config.std)
])

val_transforms = transforms.Compose([
    transforms.Resize((config.image_size, config.image_size)),
    transforms.ToTensor(),
    transforms.Normalize(config.mean, config.std)
])

# Load dataset
dataset = torchvision.datasets.ImageFolder(config.data_dir, transform=train_transforms)

# Split dataset
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size])

# Create dataloaders
train_loader = DataLoader(train_set, batch_size=config.batch_size, 
                         shuffle=True, num_workers=config.num_workers, pin_memory=True)
val_loader = DataLoader(val_set, batch_size=config.batch_size,
                       shuffle=False, num_workers=config.num_workers, pin_memory=True)
test_loader = DataLoader(test_set, batch_size=config.batch_size,
                        shuffle=False, num_workers=config.num_workers, pin_memory=True)

# ------------------------ Swin Model ------------------------
class LungCancerSwin(nn.Module):
    def __init__(self):
        super().__init__()
        self.swin = models.swin_t(weights=models.Swin_T_Weights.DEFAULT)
        
        # Freeze initial layers
        for param in self.swin.parameters():
            param.requires_grad = False
            
        # Unfreeze last 2 stages
        for block in self.swin.features[2:]:
            for param in block.parameters():
                param.requires_grad = True
                
        # Replace head
        self.in_features = self.swin.head.in_features
        self.swin.head = nn.Sequential(
            nn.Linear(self.in_features, 512),
            nn.GELU(),
            nn.Dropout(0.3),
            nn.LayerNorm(512),
            nn.Linear(512, config.num_classes)
        )
        
    def forward(self, x):
        return self.swin(x)

# ------------------------ Training Setup ------------------------
def setup_training():
    model = LungCancerSwin().to(config.device)
    criterion = nn.CrossEntropyLoss()
    optimizer = T.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=config.lr)
    scaler = T.amp.GradScaler("cuda")
    return model, criterion, optimizer, scaler

# ------------------------ Training Loop ------------------------
def train_model():
    model, criterion, optimizer, scaler = setup_training()
    best_acc = 0.0
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    
    #  Automatic Mixed Precision context manager
    for epoch in range(config.epochs):
        print(f"\nEpoch {epoch+1}/{config.epochs}")
        start_time = time.time()
        
        # Training phase
        model.train()
        train_loss, train_correct = 0.0, 0
        for images, labels in tqdm(train_loader, desc="Training"):
            images, labels = images.to(config.device), labels.to(config.device)
            
            with T.amp.autocast("cuda"):
                outputs = model(images)
                loss = criterion(outputs, labels)
            
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
            
            train_loss += loss.item() * images.size(0)
            train_correct += (outputs.argmax(1) == labels).sum().item()
        
        # Validation phase
        model.eval()
        val_loss, val_correct = 0.0, 0
        with T.no_grad():
            for images, labels in tqdm(val_loader, desc="Validating"):
                images, labels = images.to(config.device), labels.to(config.device)
                
                outputs = model(images)
                loss = criterion(outputs, labels)
                
                val_loss += loss.item() * images.size(0)
                val_correct += (outputs.argmax(1) == labels).sum().item()
        
        # Calculate metrics
        train_loss = train_loss / len(train_loader.dataset)
        train_acc = train_correct / len(train_loader.dataset)
        val_loss = val_loss / len(val_loader.dataset)
        val_acc = val_correct / len(val_loader.dataset)
        
        # Update history
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_acc)
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc)
        
        # Save best model
        if val_acc > best_acc:
            best_acc = val_acc
            T.save(model.state_dict(), os.path.join(config.save_dir, config.save_name))
            print(f" Saved new best model with val_acc: {val_acc:.4f}")
        
        print(f"Train Loss: {train_loss:.4f} | Acc: {train_acc:.4f}")
        print(f"Val Loss: {val_loss:.4f} | Acc: {val_acc:.4f}")
        print(f"Epoch time: {time.time()-start_time:.2f}s")
        
    return history, model

# ------------------------ Main Execution ------------------------
if __name__ == "__main__":
    #  Verify directories
    os.makedirs(config.save_dir, exist_ok=True)
    
    # Start training
    history, model = train_model()
    
    # Load best model for testing
    model.load_state_dict(T.load(os.path.join(config.save_dir, config.save_name)))
    
    # Final evaluation
    model.eval()
    test_correct = 0
    with T.no_grad():
        for images, labels in tqdm(test_loader, desc="Testing"):
            images, labels = images.to(config.device), labels.to(config.device)
            outputs = model(images)
            test_correct += (outputs.argmax(1) == labels).sum().item()
    
    test_acc = test_correct / len(test_loader.dataset)
    print(f"\n Final Test Accuracy: {test_acc:.4f}")

    #  Save training logs
    T.save(history, os.path.join(config.save_dir, "training_history.pt"))
```

```python
#Grok
import os
import time
import numpy as np
import torch as T
import torch.nn as nn
import torchvision
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from torchvision import models, transforms
from torch.utils.data import DataLoader, random_split
from torch.optim import Adam
from sklearn.metrics import confusion_matrix, classification_report

#  Set CUDA for Kaggle P100 GPU
device = T.device("cuda" if T.cuda.is_available() else "cpu")

# ------------------------ Hyperparameters & Config ------------------------
params = {
    'batch_size': 32,  # Suitable for P100 GPU with Swin-T
    'epochs': 30,
    'lr': 0.0003,
    'image_size': 224,  #  Swin-T uses 224x224
    'mean': [0.485, 0.456, 0.406],  # ImageNet normalization
    'std': [0.229, 0.224, 0.225],
    'device': device,
    'checkpoint_dir': "/kaggle/working",  # Kaggle working directory
    'data_dir': "/kaggle/input/pet-scans-of-lungs-98-accuracy/gdrive/MyDrive/Projects/Lung-Cancer/Data",
    'dataset_name': "Lung-PET-CT-Dx",
    'disease_names': {
        'A': 'Adenocarcinoma',
        'B': 'Small Cell Carcinoma',
        'E': 'Large Cell Carcinoma',
        'G': 'Squamous Cell Carcinoma',
    },
    'patience': 5  # Early stopping patience
}

# Ensure model save directory exists
os.makedirs(params['checkpoint_dir'], exist_ok=True)

# ------------------------ Data Preprocessing ------------------------
data_transforms = transforms.Compose([
    transforms.Resize(params['image_size']),
    transforms.ToTensor(),
    transforms.Normalize(params['mean'], params['std']),
])

dataset = torchvision.datasets.ImageFolder(params['data_dir'], transform=data_transforms)

# Splitting dataset (80% train, 10% val, 10% test)
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size
train, val, test = random_split(dataset, [train_size, val_size, test_size])

# DataLoaders with num_workers=2 for Kaggle
train_loader = DataLoader(train, batch_size=params['batch_size'], shuffle=True, num_workers=2)
val_loader = DataLoader(val, batch_size=params['batch_size'], shuffle=False, num_workers=2)
test_loader = DataLoader(test, batch_size=params['batch_size'], shuffle=False, num_workers=2)

# ------------------------ Model Definition ------------------------
class Detector(nn.Module):
    def __init__(self, base_model, n_classes, lr, name):
        super(Detector, self).__init__()
        
        self.path = os.path.join(params['checkpoint_dir'], name + '.pth')
        self.base_model = base_model

        #  Swin-T outputs 768 features
        self.in_features = self.base_model.head.in_features
        self.base_model.head = nn.Identity()  # Remove default head

        self.classifier = nn.Sequential(
            nn.Linear(self.in_features, 512),
            nn.ReLU(),
            nn.Dropout(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Dropout(),
            nn.Linear(512, n_classes),
        )

        self.criterion = nn.CrossEntropyLoss()
        self.optimizer = Adam(self.parameters(), lr)

    def forward(self, inputs):
        features = self.base_model(inputs)
        logits = self.classifier(features)
        return logits

# ------------------------ Training Function ------------------------
def fit(Model, loader, train=True):
    total_loss = 0
    total_correct = 0
    all_preds, all_labels = [], []

    if train:
        Model.train()
    else:
        Model.eval()

    #  Mixed precision for P100 optimization
    scaler = T.amp.GradScaler("cuda")

    for images, labels in tqdm(loader):
        images, labels = images.to(params['device']), labels.to(params['device'])

        if train:
            Model.optimizer.zero_grad()
        
        with T.amp.autocast("cuda"):
            predicted_logits = Model(images)
            loss = Model.criterion(predicted_logits, labels)

        if train:
            scaler.scale(loss).backward()
            scaler.step(Model.optimizer)
            scaler.update()

        total_loss += loss.cpu().detach().numpy()
        total_correct += (predicted_logits.argmax(1) == labels).sum().item()
        
        all_preds.extend(predicted_logits.argmax(1).cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    return total_loss / len(loader), total_correct / len(loader.dataset), all_preds, all_labels

# ------------------------ Model Initialization ------------------------
base = models.swin_t(weights=models.Swin_T_Weights.DEFAULT)  #  Use Swin-T
for p in base.parameters():
    p.requires_grad = False  # Freeze base model

CancerDetector = Detector(base, len(params['disease_names']), params['lr'], name='kaggle-swin-t')
Model = CancerDetector.to(params['device'])

resume = False
if resume:
    if os.path.exists(Model.path):
        checkpoint = T.load(Model.path, map_location=params['device'])
        Model.load_state_dict(checkpoint)
        print("Loaded saved model!")

history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

# ------------------------ Training Loop with Early Stopping ------------------------
start_time = time.time()
best_val_loss = float('inf')
early_stop_counter = 0

for epoch in range(params['epochs']):
    print(f'\nEpoch {epoch + 1}/{params["epochs"]}')
    
    train_loss, train_acc, _, _ = fit(Model, train_loader)
    val_loss, val_acc, val_preds, val_labels = fit(Model, val_loader, train=False)

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)

    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f}")
    print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        T.save(Model.state_dict(), Model.path)
        print(' Model Saved (Best Validation Loss)')
        early_stop_counter = 0
    else:
        early_stop_counter += 1
        print(f' Early Stopping Counter: {early_stop_counter}/{params["patience"]}')
    
    if early_stop_counter >= params['patience']:
        print(' Early Stopping Triggered!')
        break

end_time = time.time()
print(f"Total Training Time: {(end_time - start_time) / 60:.2f} minutes")

# ------------------------ Validation Confusion Matrix & Metrics ------------------------
cm = confusion_matrix(val_labels, val_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=params['disease_names'].values(), yticklabels=params['disease_names'].values())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Validation Confusion Matrix")
plt.show()

print("\nValidation Classification Report:\n", classification_report(val_labels, val_preds, target_names=params['disease_names'].values()))

# ------------------------ Test Set Evaluation ------------------------
Model.load_state_dict(T.load(Model.path, map_location=params['device']))
Model.eval()

test_loss, test_acc, test_preds, test_labels = fit(Model, test_loader, train=False)
print(f"\n Test Loss: {test_loss:.4f} | Test Accuracy: {test_acc:.4f}")

# Test Confusion Matrix
cm = confusion_matrix(test_labels, test_preds)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=params['disease_names'].values(), yticklabels=params['disease_names'].values())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Test Confusion Matrix")
plt.show()

print("\nTest Classification Report:\n", classification_report(test_labels, test_preds, target_names=params['disease_names'].values()))

# ------------------------ Plot Accuracy & Loss ------------------------
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Loss per Epoch')

plt.subplot(1, 2, 2)
plt.plot(history['train_acc'], label='Train Accuracy')
plt.plot(history['val_acc'], label='Validation Accuracy')
plt.legend()
plt.title('Accuracy per Epoch')
plt.show()

# ------------------------ Print Test Sample Paths ------------------------
test_samples = [dataset.samples[i][0] for i in test.indices[:10]]
print("First 10 Test Sample Paths:\n", test_samples)
```
